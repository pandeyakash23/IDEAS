```python
import pandas as pd

# Read the file
data = pd.read_csv('ARX_L343Q_R1_8mers.txt', sep='\t')

# Get unique 8-mers from both columns
sequences = list(data['8-mer']) + list(data['8-mer.1'])

# Duplicate the E-scores to match (same E-score for both strands)
y = list(data['E-score']) + list(data['E-score'])

print(f"Total sequences: {len(sequences)}")
print(f"First few sequences: {sequences[:5]}")
print(f"First few E-scores: {y[:5]}")
```

```python
import numpy as np
sequences = np.array(sequences)
y = np.array(y)
y = (y - y.min()) / (y.max() - y.min())
data = np.concatenate((sequences.reshape(-1,1), y.reshape(-1,1)), axis=1)
print(data.shape)
np.save('data.npy', data)
```
